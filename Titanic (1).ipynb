```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/content/gender_submission (1).csv'
df = pd.read_csv(file_path)

print("First few rows of the dataset:")
print(df.head())

print("\nBasic information about the dataset:")
print(df.info())

print("\nSummary statistics of the dataset:")
print(df.describe())

print("\nMissing values in the dataset:")
print(df.isnull().sum())

# Exploratory data analysis

# Plot the survival rate
plt.figure(figsize=(10, 6))
sns.countplot(x='Survived', data=df, palette='pastel')
plt.title('Survival Count')
plt.show()

# Since we only have 'PassengerId' and 'Survived' columns, we can explore the distribution of survival
plt.figure(figsize=(10, 6))
df['Survived'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'salmon'])
plt.title('Survival Distribution')
plt.ylabel('')
plt.show()
```
